```python
from bs4 import BeautifulSoup
import requests
import re
import urllib.request
import pandas as pd
from pandas import DataFrame
```

```python
num=1
base_url ="https://www.kijiji.ca/b-apartments-condos/canada/1+bathroom__1+half+bathrooms-1+bedroom+den__2+bedrooms-condo/page-"
page_num='1'
url_separator="/c37l0a120a27949001a29276001?ad=offering&price=__2000"
```

```python
url=base_url + page_num+ url_separator
response = requests.get(url)
soupcontent = BeautifulSoup(response.content , "html.parser")
```

```python
for post in range(15):
    p=str(post)
    base_url+ p +url_separator
    houses = soupcontent.find_all('div',{'class':'info'})
    housedata_Format=[]
```

```python
for i in houses:
    house_name = i.find('div',{'class':'title'}).get_text()
    house_price = i.find('div',{'class':'price'}).get_text()
    house_bed = i.find('div',{'class':'details'}).get_text()
    house_location = i.find('div',{'class':'location'}).get_text()
    house_description = i.find('div',{'class':'description'}).get_text()
    
    format_housename=' '.join(house_name.strip().replace('\n','').split())
    format_houseprice=' '.join(house_price.strip('').replace('$','').split())
    format_housebed=' '.join(house_bed.strip().replace('\n','').split())
    format_houselocation=' '.join(house_location.strip().replace('\n','').split())
    format_housedescription=' '.join(house_description.strip().replace('\n','').split())

    house_record= [format_housename , format_houseprice , format_housebed , format_houselocation , format_housedescription]
    housedata_Format.append(house_record)
```

```python
print(housedata_Format , '\n')
```

```python
num+=1
page_num=str(num)
df=pd.DataFrame(housedata_Format,columns=['house name','house price','house bed','house location','description'])
print(df)
```
